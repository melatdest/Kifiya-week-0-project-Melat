```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('benin-malanville.csv')
```

Check Data Info and Preview

```python
# Check the data types and null values
df.info()
```

```python
# Preview the first few rows of the data
df.head()
```

Data Cleaning: Handle missing values and non-numeric columns. For example, if your data contains non-numeric columns like Timestamp or Comments, you can exclude them from your analysis.

```python
# Drop non-numeric columns (like 'Comments', 'Timestamp')
df_cleaned = df.select_dtypes(include=['float64', 'int64'])

# Check for any missing values
df_cleaned.isnull().sum()
```

Basic EDA: Use .describe() to get statistical summaries of the data.
python
Copy code


```python
# Get a statistical summary of numeric columns
df_cleaned.describe()
```

Identify Outliers Using IQR: Calculate the Interquartile Range (IQR) to detect outliers.
python
Copy code


```python
# Calculate the 1st and 3rd quartiles (Q1 and Q3)
Q1 = df_cleaned.quantile(0.25)
Q3 = df_cleaned.quantile(0.75)
IQR = Q3 - Q1

# Identify outliers (values outside 1.5 * IQR range)
outliers = ((df_cleaned < (Q1 - 1.5 * IQR)) | (df_cleaned > (Q3 + 1.5 * IQR)))

# Print the count of outliers per column
outliers.sum()
```

Visualizations: Now, let's create some visualizations to better understand the data.
Histograms of Numeric Columns

```python
# Plot histograms for numeric columns
df_cleaned.hist(figsize=(10, 8), bins=20)
plt.tight_layout()
plt.show()
```

Boxplot for Detecting Outliers

```python
# Boxplot for each column to detect outliers
plt.figure(figsize=(10, 8))
sns.boxplot(data=df_cleaned, orient='h')
plt.title('Boxplot of Numeric Columns')
```
